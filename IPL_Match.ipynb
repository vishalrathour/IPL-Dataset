```python
# Loading all required libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Loading ipl matches dataset
ipl = pd.read_csv('IPL_Matches_2008_2022.csv')
```

```python
# loading the first five records of dataset
ipl.head()
```

```python
# looking at the number of row and columns of dataset
ipl.shape
```

```python
# getting the frequency of most man of the match awards
ipl['Player_of_Match'].value_counts()
```

```python
# getting top 10 of most man of the match awards
ipl['Player_of_Match'].value_counts()[0:10]
```

```python
# getting top 5 of most man of the match awards
ipl['Player_of_Match'].value_counts()[0:5]
```

```python
# getting top 5 of most man of the match awards
list(ipl['Player_of_Match'].value_counts()[0:5].keys())
```

```python
# by matplotlib
plt.figure(figsize=(8,5))
plt.bar(list(ipl['Player_of_Match'].value_counts()[0:5].keys()),list(ipl['Player_of_Match'].value_counts()[0:5]),color = 'c')
plt.show()
```

```python
# by seaborn
plt.figure(figsize=(8,5))
sns.barplot(x=list(ipl['Player_of_Match'].value_counts()[0:5].keys()),y=list(ipl['Player_of_Match'].value_counts()[0:5]),color = 'c')
plt.show()
```

```python
# getting the frequency of most WonBy in the match

ipl['WonBy'].value_counts()
```

```python
# finding out the number of most toss_winner team in ipl

ipl['TossWinner'].value_counts()
```

```python
# Extracting the records where a team won batting first and win by runs

batting_first = ipl[ipl['WonBy']=='Runs']
```

```python
batting_first
```

```python
# Extracting the records where a team won batting second and win by wickets

batting_second = ipl[ipl['WonBy']=='Wickets']
```

```python
batting_second
```

```python
# making a histogram 
plt.figure(figsize=(5,7))
plt.hist(batting_first['WonBy'])
plt.title('Distributions of Runs')
plt.show()
```

```python
batting_first['WonBy'].keys()
```

```python
# finding out the number of the wins w.r.t. each team after batting first
batting_first['WinningTeam'].value_counts()
```

```python
plt.figure(figsize=(8,6))
plt.bar(list(batting_first['WinningTeam'].value_counts()[0:3].keys())
        , list(batting_first['WinningTeam'].value_counts()[0:3]) , color = ['red','blue','orange'])
plt.title('Win in first time batting')
plt.show()
```

```python
x = [1,2,3,4,5,62,4,5,6,7,8,9,5,1,1,21,1,2,12,]
y = x*3
sns.histplot(y)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.pie(list(batting_first['WinningTeam'].value_counts()[0:10]),
        labels = list(batting_first['WinningTeam'].value_counts()[0:10].keys()),
       autopct='%0.2f%%')
plt.title('Win in first time batting', color = 'b')
plt.show()
```
